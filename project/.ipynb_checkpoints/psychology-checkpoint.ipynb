```python
import pyodbc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#connect to mssql
server = '128.1.99.58'
database = 'ID_XYLITOL_WHITE'
username = ''
password = ''

cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER='+server+';DATABASE='+database+';UID='+username+';PWD='+ password)
cursor = cnxn.cursor()
```

```python
# 치매를 진단받은 환자 선택 (동일한 질병에 대해 여러 번 병원을 방문한 기록이 모두 포함)

AD='''
select st.person_id, st.condition_start_date, nd.gender_source_value, nd.year_of_birth
from (SELECT co.person_id, co.condition_concept_id, co.condition_start_date
  FROM [CDMPv531].[dbo].CONDITION_OCCURRENCE co
  where co.condition_concept_id in ('378419', '4220313', '4218017')) st
  join (select person_id, gender_source_value, year_of_birth FROM [CDMPv531].[dbo].person) nd on st.person_id=nd.person_id
'''
AD = pd.read_sql_query(AD, cnxn)

AD_part=pd.DataFrame(AD.groupby('person_id').condition_start_date.min())
AD_part['person_id'] = AD_part.index
AD_part=AD_part.reset_index(drop=True)

AD
```

```python
# 각 환자가 치매를 처음 진단 받은 날짜가 속한 행만 선택
## 각 환자의 나이를 계산

AD_real=pd.merge(left=AD_part, right=AD, how='left', on=['person_id','condition_start_date'], sort=False).drop_duplicates()
def age(row):
    return int(row['condition_start_date'][0:4])-row['year_of_birth']
AD_real['age']=AD_real.apply(age, axis=1)
AD_real
```

```python
# 알츠하이머 치매 환자의 성비를 시각화하기 위한 파이 차트

group_names = ['Male', 'Female']
group_sizes = [list(AD_real['gender_source_value']).count('M'), list(AD_real['gender_source_value']).count('F')]
group_colors = ['yellowgreen', 'lightskyblue']
group_explodes = (0.1, 0) # explode 1st slice

plt.pie(group_sizes, 
        explode=group_explodes, 
        labels=group_names, 
        colors=group_colors, 
        autopct='%1.2f%%', # second decimal place
        shadow=True, 
        startangle=90,
        textprops={'fontsize': 14}) # text font size
plt.axis('equal') #  equal length of X and Y axis
plt.title('Pie Chart for gender proportion of AD patients', fontsize=20)
plt.show()
```

```python
# 알츠하이머 치매 환자의 진단 시점 나이 분포를 시각화하기 위한 histogram

import numpy as np
data=list(AD_real['age'])
bins = np.arange(-100, 100, 5)
plt.xlim([min(data)-5, max(data)+5])

plt.hist(data, bins=bins, alpha=0.5)
plt.title('Age distribution at the time of diagnosis in patients with AD')
plt.xlabel('Age at the time of diagnosis of AD')
plt.ylabel('count')

plt.show()
```
